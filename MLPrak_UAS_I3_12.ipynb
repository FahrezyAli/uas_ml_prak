```python
import pandas as pd
import numpy as np
import nltk
nltk.download('punkt')
nltk.download('wordnet')
nltk.download('stopwords')
nltk.download('omw-1.4')
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem.wordnet import WordNetLemmatizer
from ast import literal_eval
```

```python
# Mengambil data dari file csv
data = pd.read_csv('Hotel_Reviews.csv')
```

```python
# Preprocessing

# Membuat map untuk mengubah negara menjadi kode negara
country_codes = {
    'Netherlands': 'NL',
    'United Kingdom': 'UK',
    'France': 'FR',
    'Spain': 'ES',
    'Italy': 'IT',
    'Austria': 'AT'
}

# Mengubah neggara menjadi kode negara
for country, code in country_codes.items():
    data.Hotel_Address = data.Hotel_Address.str.replace(country, code)
```

```python
# Mengambil strl terakhir dari Hotel_Address yang mana merupakah negara dari hotel tersebut
# Mmemasukannya ke dalam kolom baru yaitu Countries
data["Countries"] = data.Hotel_Address.apply(lambda x: x.split(' ')[-1])
```

```python
# Menghapus kolom yang tidak diperlukan
data.drop(['Additional_Number_of_Scoring', 
           'Review_Date', 'Reviewer_Nationality', 
           'Negative_Review', 'Review_Total_Negative_Word_Counts', 
           'Total_Number_of_Reviews', 'Positive_Review', 
           'Review_Total_Positive_Word_Counts', 
           'Total_Number_of_Reviews_Reviewer_Has_Given', 'Reviewer_Score', 
           'days_since_review', 'lat', 'lng'], 
          axis=1, inplace=True)
```

```python
# Mendapatkan tags demi implementasi sebelum di impute pada tahap selanjutnya
def get_unique_values(df, column):
    # Extract unique tags from the specified column in the DataFrame
    unique_values = set()
    df[column].apply(lambda x: unique_values.update(x.split(',')))
    return list(unique_values)

# Get unique tags from the DataFrame
unique_tags = get_unique_values(data, "Tags")
```

```python
#  Membuat fungsi impute yang akan digunakan untuk menggabungkan list menjadi string

def impute(column):
    column = column[0]
    if (type(column) != list):
        return "".join(literal_eval(column))
    else:
        return column
```

```python
# Apply fungsi impute ke kolom Tags
data["Tags"] = data[["Tags"]].apply(impute, axis=1)
data.head()
```

```python
# Analisa Data

# Membuat fungsi system recommender
def recommender(location, description):
    # Dividing the texts into small tokens (sentences into words)
    word_tokenize(description)
    # applying stopwords
    stop_words = stopwords.words('english')
    # applying lemmatization
    lemm = WordNetLemmatizer()
    # connecting packages and filtering
    filtered = {word for word in description if not word in stop_words}
    filtered_set = set()
    for fs in filtered:
        filtered_set.add(lemm.lemmatize(fs))
    # creating a variable that takes in the location and returns the following features
    country = data[data['Countries'] == location]
    country = country.set_index(np.arange(country.shape[0]))
    list1 = []; list2 = []; cos = []
    for i in range(country.shape[0]):
        temp_token = word_tokenize(country["Tags"][i])
        temp_set = [word for word in temp_token if word not in stop_words]
        temp2_set = set()
        for s in temp_set:
            temp2_set.add(lemm.lemmatize(s))
        vector = temp2_set.intersection(filtered_set)
        cos.append(len(vector))
    # applying cosine similarity
    country['similarity'] = cos
    country = country.sort_values(by='similarity', ascending=False)
    country.drop_duplicates(subset='Hotel_Name', keep='first', inplace=True)
    country.sort_values('Average_Score', ascending=False, inplace=True)
    country.reset_index(inplace=True)

    return country[["Hotel_Name", "Average_Score", "Hotel_Address"]]
```

```python
# Mencoba nya pada satu contoh kasus
recommender('UK','Business trip')
```

```python
# Perancangan System

import re
import tkinter as tk
from tkinter import ttk

# Fungsi untuk menampilkan UI system recommender
# Fungsi submit
def submit_action():
    dropdown1_value = dropdown1.get()
    country = country_to_code(dropdown1_value)

    selected_tags_indices = listbox.curselection()
    selected_tags_list = [listbox.get(i) for i in selected_tags_indices]
    selected_tags = ' '.join(selected_tags_list)

    data = recommender(country, selected_tags)

    text_widget.delete(1.0, tk.END)
    text_widget.insert(tk.END, data.to_string(index=False))

# Define fungsi untuk merubah negara menjadi kode negara
def country_to_code(country):
    return country_codes.get(country, '')

# Define fungsi untuk mengupdate listbox saat user menggunakan search box
def update_listbox(search_query):
    # Clear the listbox
    listbox.delete(0, tk.END)
    
    # Filter tags based on the search query
    filtered_tags = [tag for tag in unique_tags if search_query.lower() in tag.lower()]
    
    # Insert filtered tags into the listbox
    for tag in filtered_tags:
        tag_stripped = re.sub(re.compile(r'^\W*|\W*$'), '', tag)
        listbox.insert(tk.END, tag_stripped)

# Main Window
root = tk.Tk()
root.title("Hotel Recommender System")

# Dropdown negara
label1 = ttk.Label(root, text="Select Country")
label1.grid(column=0, row=0, padx=10, pady=10)
dropdown1 = ttk.Combobox(root, values=list(country_codes.keys()))
dropdown1.grid(column=1, row=0, padx=10, pady=10)

# Search Tags
label2 = ttk.Label(root, text="Search Tags:")
label2.grid(column=0, row=1, padx=10, pady=10)
search_entry = ttk.Entry(root)
search_entry.grid(column=1, row=1, padx=10, pady=10)

# Listbox Tags
label3 = ttk.Label(root, text="Select Tags:")
label3.grid(column=0, row=2, padx=10, pady=10)
listbox = tk.Listbox(root, selectmode=tk.MULTIPLE, width=50)
for tag in unique_tags:
    tag_stripped = re.sub(re.compile(r'^\W*|\W*$'), '', tag)
    listbox.insert(tk.END, tag_stripped)
listbox.grid(column=1, row=2, padx=10, pady=10)

# Update listbox saat user mengetikkan sesuatu
search_entry.bind("<KeyRelease>", lambda event: update_listbox(search_entry.get()))

# Tombol submit
submit_button = ttk.Button(root, text="Submit", command=submit_action)
submit_button.grid(column=0, row=3, columnspan=2, pady=20)

# Window untuk menampilkan hasil
text_widget = tk.Text(root, wrap="none", width=150, height=20)
text_widget.grid(column=0, row=4, columnspan=2, padx=10, pady=10)

# Mulai UI
root.mainloop()
```
